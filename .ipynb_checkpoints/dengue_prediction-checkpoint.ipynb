# DengAI: Predicting Disease Spread
Your goal is to predict the total_cases label for each (city, year, weekofyear) in the test set. There are two cities, San Juan and Iquitos, with test data for each city spanning 5 and 3 years respectively. 

The data for each city have been concatenated along with a city column indicating the source: sj for San Juan and iq for Iquitos. The test set is a pure future hold-out, meaning the test data are sequential and non-overlapping with any of the training data. 

## Data description
** City and date indicators **
- city – City abbreviations: sj for San Juan and iq for Iquitos
- week_start_date – Date given in yyyy-mm-dd format


** NOAA's GHCN daily climate data weather station measurements **


- station_max_temp_c – Maximum temperature
- station_min_temp_c – Minimum temperature
- station_avg_temp_c – Average temperature
- station_precip_mm – Total precipitation
- station_diur_temp_rng_c – Diurnal temperature range


** PERSIANN satellite precipitation measurements (0.25x0.25 degree scale)**


- precipitation_amt_mm – Total precipitation


** NOAA's NCEP Climate Forecast System Reanalysis measurements (0.5x0.5 degree scale) **


- reanalysis_sat_precip_amt_mm – Total precipitation
- reanalysis_dew_point_temp_k – Mean dew point temperature
- reanalysis_air_temp_k – Mean air temperature
- reanalysis_relative_humidity_percent – Mean relative humidity
- reanalysis_specific_humidity_g_per_kg – Mean specific humidity
- reanalysis_precip_amt_kg_per_m2 – Total precipitation
- reanalysis_max_air_temp_k – Maximum air temperature
- reanalysis_min_air_temp_k – Minimum air temperature
- reanalysis_avg_temp_k – Average air temperature
- reanalysis_tdtr_k – Diurnal temperature range


** Satellite vegetation - Normalized difference vegetation index (NDVI) - NOAA's CDR Normalized Difference Vegetation Index (0.5x0.5 degree scale) measurements **


- ndvi_se – Pixel southeast of city centroid
- ndvi_sw – Pixel southwest of city centroid
- ndvi_ne – Pixel northeast of city centroid
- ndvi_nw – Pixel northwest of city centroid

# Feature engineering

Most of the temperature reading's are in kelvin , so we will convert celsius reading to kelvin

```r
celsius_columns = colnames(train)[endsWith(x = colnames(train),"_c")]
```

```r
train[,celsius_columns] <- apply(train[,celsius_columns],2,function(x) celsius.to.kelvin(T.celsius = x))
```

```r
# quarter
train$quarter <- quarters(train$week_start_date)
```

```r
# month
train$month <- months.Date(train$week_start_date)
```

Rolling mean

```r
roll_mean_vars <- colnames(train)[!colnames(train) %in% c(exclude,"city","year_week","total_cases","ndvi_se","ndvi_sw","ndvi_ne","ndvi_nw")]
```

```r
roll_mean_vars_3 <- paste(roll_mean_vars, "roll_mean_3", sep = "_")
```

```r
roll_mean_vars_4 <- paste(roll_mean_vars, "roll_mean_4", sep = "_")
```

```r
train_dimension =  dim(train)
```

```r
train[,roll_mean_vars_3] <- NA
train[,roll_mean_vars_4] <- NA
```

```r
train[3:train_dimension[1],roll_mean_vars_3] <- apply(train[,roll_mean_vars],2,function(x) rollmean(x,k = 3))
```

```r
train[4:train_dimension[1],roll_mean_vars_4] <- apply(train[,roll_mean_vars],2,function(x) rollmean(x,k = 4))
```

Cumulative sum

```r
cumsum_vars <- colnames(train)[!colnames(train) %in% c(exclude,"city","year_week","total_cases","ndvi_se","ndvi_sw","ndvi_ne","ndvi_nw")]
```

```r
cumsum_vars_4 <- paste(roll_mean_vars, "cumsum_4", sep = "_")
```

```r
train[,cumsum_vars_4] <- NA
```

```r
train[4:train_dimension[1],cumsum_vars_4] <- apply(train[,cumsum_vars],2,function(x) rollapply(x, 4, sum))
```
